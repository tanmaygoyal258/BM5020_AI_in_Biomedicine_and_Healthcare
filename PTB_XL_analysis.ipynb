```python
import pandas as pd
```

```python
df = pd.read_csv('data/PTB_XL/ptbxl_database.csv')
```

```python
df
```

```python
df['sex'].value_counts(normalize = True)
# 0 is male
# 1 is female
```

```python
df['filename'] = df['filename_lr'].apply(lambda x : x.split('/')[-1].split('_')[0])
```

```python
df.columns
```

```python
df_subset = df[['filename','age','sex']]
```

```python
labels = pd.read_csv("data/PTB_XL/labels_with_final_mapping.csv" , dtype = {'patient_id' : str})
labels.rename(columns = {'patient_id':'filename'}, inplace = True)
```

```python
meta_df = pd.merge(labels , df_subset , on = 'filename' , how = 'left')
```

```python
meta_df
```

```python
diagnosis = ['SNR','AF', 'IAVB' , 'LBBB' , 'RBBB' , 'PAC' , 'PVC' , 'STD' , 'STE']

for d in diagnosis:
    subset = meta_df[meta_df[d] == 1]
    print("\\textbf{{{}}}&".format(d))
    print("{:.0f} ({:.2f})&".format(len(subset) , len(subset)/len(meta_df)*100))
    male_count = len(subset[subset['sex'] == 0])
    print("{:.0f} ({:.2f})&".format(male_count , male_count/len(subset)*100))
    print("{:.2f} ({:.2f})\\\\\\hline".format(subset['age'].mean() , subset['age'].std()))

    
```
