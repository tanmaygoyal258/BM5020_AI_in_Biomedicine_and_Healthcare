```python
import pandas as pd
import numpy as np
```

```python
label_csv_path = "/Users/tanmaygoyal/Desktop/Assignments and Events/Biomedicine/ECG_Project/data/PTB_XL/labels.csv"
label_csv = pd.read_csv(label_csv_path , dtype={'patient_id':str})
```

```python
label_csv
```

```python
labels_to_remove = ['ISCIN','ISCIL','DIG','ISCLA','RVH','LNGQT','EL','ISCAN','SEHYP','STACH','RAO/RAE']
```

```python
label_csv.drop(labels_to_remove , axis = 1 , inplace = True)
```

```python
label_csv
```

```python
potential_diseases = [c for c in label_csv.columns if c != 'patient_id']
potential_diseases
```

```python
patient_ids_to_remove = label_csv[label_csv[potential_diseases].sum(axis = 1) == 0]['patient_id']
```

```python
remove_data_path = "/Users/tanmaygoyal/Desktop/Assignments and Events/Biomedicine/ECG_Project/data/PTB_XL/removed_data.txt"
with open(remove_data_path , 'w') as f:
    for p in patient_ids_to_remove:
        f.write(p + '.mat\n')
```

```python
label_csv.drop(label_csv[label_csv['patient_id'].isin(patient_ids_to_remove)].index , axis = 0 , inplace = True)
```

```python
label_csv.shape
```

```python
label_mapping = pd.read_csv("/Users/tanmaygoyal/Desktop/Assignments and Events/Biomedicine/ECG_Project/data/PTB_XL/label_mapping.csv")
```

```python
ptb_xl_disease = label_mapping['PTB_XL'].tolist()
cpsc_disease = label_mapping['CPSC'].tolist()
```

```python
final_label_csv = label_csv['patient_id'].to_frame()

unique_cpsc_diseases = list(set(cpsc_disease))

for disease in unique_cpsc_diseases:
    final_label_csv[disease] = 0
    for ptbxl,cpsc in zip(ptb_xl_disease,cpsc_disease):
        if cpsc == disease and ptbxl in label_csv.columns:
            all_with_label1 = label_csv[label_csv[ptbxl] == 1]['patient_id'].tolist()  
            final_label_csv.loc[final_label_csv['patient_id'].isin(all_with_label1) ,disease] = 1    
```
